```python
import urllib.request
from bs4 import BeautifulSoup
from newspaper import Article
import pandas as pd
import stockstats
import csv
import nltk
import time
import bs4
import re
```

```python
def stat_list():
    url = "https://www.moneycontrol.com/stocks/marketstats/indexcomp.php?optex=NSE&opttopic=indexcomp&index=9"
    # Open the URL as Browser, not as python urllib
    page=urllib.request.Request(url,headers={'User-Agent': 'Mozilla/5.0'}) 
    infile=urllib.request.urlopen(page).read()
    soup = BeautifulSoup(infile,'lxml')
    section = soup.find("table", {"class": "tbldata14 bdrtpg"})

    test_string=""
    for row in section.find_all('tr'):

        unwanted = row.find('div')
        if isinstance(unwanted, bs4.element.Tag):
            unwanted.extract() #to remove sub-div present (if any) 
        test_string+=row.text
        test_string+="******" # to seperate row values
    test_string=test_string.strip()
    string = re.sub(r"\s{3,}", "\n", test_string)
    l1=string.split("******") # getting table rows
    for i in range (len(l1)):
        l1[i]=l1[i].split("\n") #getting table data values

    l1.pop()
    l1.pop(0)  #removing empty strings present in the list
    for i in range (len(l1)):
        l1[i].pop()
        l1[i].pop(0)

    return l1
```

```python
initial_list = stat_list()
old_list = stat_list()

total_stock_list=[[[]]] #3-D List to store the values(opening and closing) for all the companys seperately
for i in range (len(old_list)):
        total_stock_list.append([[float(re.sub(r",","",initial_list[i][2])),float(re.sub(r",","",old_list[i][2]))]])

total_stock_list.pop(0)   #removing the empty element

start_time=time.clock()
five_min=5*60
print("Starting Script!\n")
while(time.clock()-start_time<=five_min):
    if(int(time.clock()-start_time)%30==0): #refreshin at intervals of 30sec
        diff_list = []
        print("\n\nUpdating data!\n")
        new_list = stat_list()
        
        
        for i in range (len(new_list)):
            diff_list.append(abs(float(new_list[i][4])-float(old_list[i][4])))
            total_stock_list[i].append([total_stock_list[i][-1][1],float(re.sub(r",","",new_list[i][2]))]) #adding new opening and closing values
        
        for i in range (len(diff_list)):
            if(diff_list[i]>=2):
                print("\nAlert! More than 2% difference in:\n",new_list[i][0]," from",old_list[i][4]," to",new_list[i][4])
                old_list=new_list #updating old list to avoid recursion in printing


print("\nSuccessfully Completed\n")
```

```python
def apply_indicator(choice,data_frame):
    stock_df = stockstats.StockDataFrame.retype(data_frame) # creating a stock data frame using stockstats
    if(choice==1):
        data_frame['rsi_6']=stock_df['rsi_6']
        print("\n",data_frame)
    
    elif(choice==2):
        data_frame['macd']=stock_df['macd']
        print("\n",data_frame)
        
    elif(choice==3):
        data_frame['change']=stock_df['change']
        print("\n",data_frame)
        
    else:
        print("\nPlease Enter a valid choice!\n")

```
